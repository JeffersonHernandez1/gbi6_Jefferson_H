```python
# Se importa el módulo alelos y se contruye 100 generaciones con una población de 15 individuos y una probabilidad de ocurrencia del alelo dominante de 0.56
import alelos as al

N = 100
for i in range(N):
    print(i, al.build_population(15, 0.56))
```

```python
# Se contea el par de alelos de las generaciones obtenidas en el item 1
N = 100
for i in range(N):
    print(i, al.compute_frequencies(al.build_population(15, 0.56)))
```

```python
# Se reproduce la población a partir de las generaciones anteriormente obtenidas
N = 100
for i in range(N):
    print(i, al.reproduce_population(al.build_population(15, 0.56)))
```
